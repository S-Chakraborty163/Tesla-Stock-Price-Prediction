```python
tesla['Date'] = pd.to_datetime(tesla['Date'])
tesla = tesla.set_index('Date')
```

```python
tesla.resample('YE').max()['Open'].plot()
plt.title('Yearly Maximum Opening Price')
plt.ylabel('Price ($)')
plt.show()
```

```python
tesla.resample('QS').max()['High'].plot()
```

```python
plt.show()
```

```python
tesla['Open'].resample('BA').mean().plot(kind='bar')
plt.show()
```

```python
tesla.resample('BQS').max()['High'].plot()
plt.show()
```

```python
tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(35,6))
plt.show()
```
